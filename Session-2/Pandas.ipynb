```python
import seaborn as sns
df = sns.load_dataset('titanic')
```

```python
df.head()
```

```python
df.shape
```

```python
numeric_data = df.select_dtypes(include=[np.number])
cat_data = df.select_dtypes(exclude=[np.number])
```

```python
numeric_data.head()
```

```python
cat_data.head()
```

```python
def chk(col):
    if col == 'S':
        return 1
    elif col == 'C':
        return 2
    else:
        return 3
```

```python
cat_data['embarked']=cat_data['embarked'].apply(chk).head()
```

```python
cat_data.head()
```

```python
cat_data.describe(include='all')
```

```python
cat_data=cat_data.drop(['embarked'],axis=1)
```
